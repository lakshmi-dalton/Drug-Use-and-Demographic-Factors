We must initialize two separate dataframes using the datasets referenced above. Given the data, we list the relevant variables from each dataset that are most beneficial to our analysis. All other variables are dropped.

Notice the ambiguity of the column names. We must rename them to be more descriptive.

```python
new_col_names = ['marijuana_use', 'marijuana_use_in_year', 'marijuana_use_in_month', 'cocaine_use', 'cocaine_use_in_year', 'cocaine_use_in_month', 'crack_use', 'crack_use_in_year', 'crack_use_in_month', 'heroin_use', 'heroin_use_in_year', 'heroin_use_in_month', 'meth_use', 'meth_use_in_year', 'meth_use_in_month', 'age', 'sexual_identity', 'gender', 'marital_status', 'highest_education', 'partner_in_military', 'parent_in_military', 'weekly_work_situation', 'has_medicare', 'total_family_income', 'poverty_level', 'county_size']
nsduh20_df.columns = new_col_names
nsduh21_df.columns = new_col_names
```

To further clean the data, we drop the rows that have NaN values. This must be done to account for the unreliability among responses with invalid data.

```python
nsduh20_df = nsduh20_df.dropna()
nsduh21_df = nsduh21_df.dropna()
```

According to the codebook, each response falls within a valid range or collection of integers. For example, any value under the variable `marijuana_use_in_month` must fall within the set [1, 2, 3, ..., 30, 85, 91, 93, 94, 97, 98], where 85 = 'BAD DATA', 91 = 'NEVER USED MARIJUANA', etc..

For any value that fails to be within the set of integers (i.e. type of responses) we most want, we replace it with NaN.

```python
# Initializing a list of variables and their corresponding valid responses
mask_list = [('marijuana_use', range(1, 3)), \
             ('cocaine_use', range(1, 3)), \
             ('crack_use', range(1, 3)), \
             ('meth_use', range(1, 3)), \
             ('heroin_use', range(1, 3)), \
             ('marijuana_use_in_year', list(range(366)) + [991, 993]) , \
             ('cocaine_use_in_year', list(range(366)) + [991, 993]) , \
             ('crack_use_in_year', list(range(366)) + [991, 993]) , \
             ('meth_use_in_year', list(range(366)) + [991, 993]) , \
             ('heroin_use_in_year', list(range(366)) + [991, 993]) , \
             ('marijuana_use_in_month', list(range(1,31)) + [91, 93]), \
             ('cocaine_use_in_month', list(range(1,31)) + [91, 93]), \
             ('crack_use_in_month', list(range(1,31)) + [91, 93]), \
             ('meth_use_in_month', list(range(1,31)) + [91, 93]), \
             ('heroin_use_in_month', list(range(1,31)) + [91, 93]), \
             ('sexual_identity', range(1, 4)), \
             ('gender', range(1,3)), \
             ('marital_status', range(1,5)), \
             ('highest_education', range(1, 12)), \
             ('partner_in_military', range(1,3)), \
             ('parent_in_military', range(1,3)), \
             ('weekly_work_situation', range(1,9)), \
             ('has_medicare', range(1,3)), \
             ('total_family_income', range(1, 7)), \
             ('poverty_level', list(range(1, 4)) + ['.']), \
             ('county_size', range(1, 4))]

for obs in mask_list:
    nsduh20_df[obs[0]].mask(~nsduh20_df[obs[0]].isin(obs[1]), inplace=True)
    nsduh21_df[obs[0]].mask(~nsduh21_df[obs[0]].isin(obs[1]), inplace=True)
    
# Accounting for special case in which the datasets differed in methodology
nsduh20_df['age'].mask(~nsduh20_df['age'].isin(range(1, 18)), inplace=True)
nsduh21_df['age'].mask(~nsduh21_df['age'].isin(range(1, 12)), inplace=True)
```

Again, we drop the rows that have NaN values. However, we only drop a row if it contains *all* NaN values. We believe that rows with some NaN values are still able to provide us with relevant information. For example, people may have used cocaine but not crack.

Note: The difference between a NaN before and after the masking (above) is that a NaN before represents invalid data whereas a NaN after respresents data that simply went beyond the type of responses we want.

```python
nsduh20_df = nsduh20_df.dropna(how='all')
nsduh21_df = nsduh21_df.dropna(how='all')
```

Last, we must standardize the data. To do so, we change values 91, 93, 991, and 993 to 0 and convert the substance variables (i.e. those with respect to year and month) to frequencies.

```python
# Replacing 91, 93, 991, and 993 with 0 in both datasets
month_vars = ['marijuana_use_in_month', 'cocaine_use_in_month', 'crack_use_in_month', 'meth_use_in_month', 
            'heroin_use_in_month']
year_vars = ['marijuana_use_in_year', 'cocaine_use_in_year', 'crack_use_in_year', 'meth_use_in_year', 
            'heroin_use_in_year']

for col in month_vars:
    nsduh20_df[col].replace(91, 0, inplace=True)
    nsduh20_df[col].replace(93, 0, inplace=True)
    nsduh21_df[col].replace(91, 0, inplace=True)
    nsduh21_df[col].replace(93, 0, inplace=True)
    
for col in year_vars:
    nsduh20_df[col].replace(991, 0, inplace=True)
    nsduh20_df[col].replace(993, 0, inplace=True)
    nsduh21_df[col].replace(991, 0, inplace=True)
    nsduh21_df[col].replace(993, 0, inplace=True)
    
# Converting to frequencies
nsduh20_df[month_vars] = nsduh20_df[month_vars].apply(lambda x: x / 30)
nsduh21_df[month_vars] = nsduh21_df[month_vars].apply(lambda x: x / 30)
nsduh20_df[year_vars] = nsduh20_df[year_vars].apply(lambda x: x / 365)
nsduh21_df[year_vars] = nsduh21_df[year_vars].apply(lambda x: x / 365)
```

# Data Analysis & Results (EDA)

For the EDA portion of our analysis, we isolate each non-substance variable and do the following: analyze its distribution, highlight any outliers, and discuss its relationship with the substance variables.

Note: Since we listed a number of non-substance variables, the amount of code needed to compare each one to every substance variable can be substantially large. In order to avoid this, we instead compare some non-substance variables with the average substance use in a month and year (as opposed to all 10 variables), computed below:

```python
# Initializing list of substance variables in both datasets
substance_variables = ['marijuana_use_in_year', 'marijuana_use_in_month', 'cocaine_use_in_year', 'cocaine_use_in_month', \
                       'crack_use_in_year', 'crack_use_in_month', 'heroin_use_in_year', 'heroin_use_in_month', \
                       'meth_use_in_year', 'meth_use_in_month']

# Dropping any unwanted responses
substance_20_df = nsduh20_df[substance_variables].dropna(how='any')
substance_21_df = nsduh21_df[substance_variables].dropna(how='any')

# Summing each observation's (i.e. respondent's) number of days for which each substance was used in a year and month
substance_20_df['avg_substance_use_in_year'] = substance_20_df['marijuana_use_in_year'] + substance_20_df['cocaine_use_in_year'] \
                                             + substance_20_df['crack_use_in_year'] + substance_20_df['heroin_use_in_year'] \
                                             + substance_20_df['meth_use_in_year']
substance_20_df['avg_substance_use_in_month'] = substance_20_df['marijuana_use_in_month'] + substance_20_df['cocaine_use_in_month'] \
                                             + substance_20_df['crack_use_in_month'] + substance_20_df['heroin_use_in_month'] \
                                             + substance_20_df['meth_use_in_month']
substance_21_df['avg_substance_use_in_year'] = substance_21_df['marijuana_use_in_year'] + substance_21_df['cocaine_use_in_year'] \
                                             + substance_21_df['crack_use_in_year'] + substance_21_df['heroin_use_in_year'] \
                                             + substance_21_df['meth_use_in_year']
substance_21_df['avg_substance_use_in_month'] = substance_21_df['marijuana_use_in_month'] + substance_21_df['cocaine_use_in_month'] \
                                             + substance_21_df['crack_use_in_month'] + substance_21_df['heroin_use_in_month'] \
                                             + substance_21_df['meth_use_in_month']

# Computing the average number of days for which each substance was used in a year and month
substance_20_df['avg_substance_use_in_year'] = substance_20_df['avg_substance_use_in_year'].apply(lambda x: x / 5)
substance_20_df['avg_substance_use_in_month'] = substance_20_df['avg_substance_use_in_month'].apply(lambda x: x / 5)
substance_21_df['avg_substance_use_in_year'] = substance_21_df['avg_substance_use_in_year'].apply(lambda x: x / 5)
substance_21_df['avg_substance_use_in_month'] = substance_21_df['avg_substance_use_in_month'].apply(lambda x: x / 5)
```

We also define a helpful function below to increase readability and reduce the overall amount of code.

```python
# Def'n: Computes the frequencies for a particular non-substance variable
def get_frequency_df(column_name):
    counts_20 = nsduh20_df[column_name].value_counts()
    counts_21 = nsduh21_df[column_name].value_counts()
    tot_count_20 = np.sum(counts_20)
    tot_count_21 = np.sum(counts_21)
    freq_20 = []
    freq_21 = []
    
    for i in range(len(counts_20.values)):
        freq_20.append([counts_20.index[i], counts_20.values[i] / tot_count_20])
        
    for i in range(len(counts_21.values)):
        freq_21.append([counts_21.index[i], counts_21.values[i] / tot_count_21])
        
    freq_20_df = pd.DataFrame(data=freq_20, columns=[column_name + '_response_value', 'frequency'])
    freq_21_df = pd.DataFrame(data=freq_21, columns=[column_name + '_response_value', 'frequency'])
    
    return freq_20_df, freq_21_df
```

```python
# -----------------------------
# --------- Constants ---------
# -----------------------------

p_alpha = 0.05

avg_yearly_no_zero_20_df = substance_20_df.loc[substance_20_df['avg_substance_use_in_year'] != 0, ['avg_substance_use_in_year']]
avg_yearly_no_zero_21_df = substance_21_df.loc[substance_21_df['avg_substance_use_in_year'] != 0, ['avg_substance_use_in_year']]

# -----------------------------
# --------- Functions ---------
# -----------------------------

# Def'n: An alternative to one-way ANOVA that does not assume equal variances
def welch_anova_test(dependent, independent, df):
    return pg.welch_anova(dv=dependent, between=independent, data=df)

# Def'n: A non-parametric post hoc analysis that compares different groups
def games_howell_test(dependent, independent, df):
    return pg.pairwise_gameshowell(dv=dependent, between=independent, data=df)

# Def'n: Returns which variable category has the greatest mean
def get_greatest_mean_in_gh(df):
    max_one = df['mean(A)'].max()
    max_two = df['mean(B)'].max()
    
    ret = df[df['mean(A)'] == max_one]['A'] if max_one > max_two else df[df['mean(B)'] == max_two]['B']
    
    return ret.unique()[0]

# Def'n: Returns a tuple of the two categories with the largest
#        difference in means
def get_greatest_diff_in_gh(df):
    curr_greatest_diff = -999
    mean_A = None
    mean_B = None
    
    for ind in df.index:
        curr_diff = np.abs(df['mean(A)'][ind] - df['mean(B)'][ind])
        
        if curr_diff > curr_greatest_diff:
            curr_greatest_diff = curr_diff
            mean_A = df['mean(A)'][ind]
            mean_B = df['mean(B)'][ind]
        
    return (df[df['mean(A)'] == mean_A]['A'].unique()[0], df[df['mean(B)'] == mean_B]['B'].unique()[0])

# Def'n: Returns a dictionary where key = 'category' and value = 
#        'corresponding averages of substance use in a year'
def get_avg_by_category(column_name):
    df_20 = pd.DataFrame().assign(avg_substance_use_in_year = avg_yearly_no_zero_20_df, \
                                  column_name = nsduh20_df[column_name].dropna(how='any'))
    df_21 = pd.DataFrame().assign(avg_substance_use_in_year = avg_yearly_no_zero_21_df, \
                                  column_name = nsduh21_df[column_name].dropna(how='any'))
    
    df_20.dropna(how='any', inplace=True)
    df_21.dropna(how='any', inplace=True)
    
    df_20.rename(columns={'column_name': column_name}, inplace=True)
    df_21.rename(columns={'column_name': column_name}, inplace=True)
        
    groups_20 = df_20.groupby(column_name)
    groups_21 = df_21.groupby(column_name)
    
    groups_k_20 = {}
    groups_k_21 = {}
    
    list_tmp_20 = list(df_20[column_name].unique())
    list_tmp_21 = list(df_21[column_name].unique())
    
    for category in list_tmp_20:
        groups_k_20[category] = groups_20.get_group(category)['avg_substance_use_in_year']
    for category in list_tmp_21:
        groups_k_21[category] = groups_21.get_group(category)['avg_substance_use_in_year']
    
    return df_20, groups_k_20, df_21, groups_k_21
```

For our inferential analyses, though we have computed values for average substance use in a year and month, we believe that showing a relationship between the former and our non-substance variables is more valuable. In addition, we exclude observations in which the average substance use in a year is zero (see *avg_yearly_no_zero_df* in the above cell). This is because most of the yearly averages, when split by the unique values of a non-substance variable, weigh heavily towards zero, skewing the distributions significantly.

For Welch's ANOVA, the null hypothesis is that there is no statistically, significant difference in the means of three or more groups. Since the p-values are less than *p_alpha*, we reject the null hypothesis. Notably, the percentage of male respondents is higher than that of female respondents. Given males are more likely to be victims of substance abuse, we have to consider this disparity while processing the data.
